```python
from bs4 import BeautifulSoup
import requests
```

```python
dineout= requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
dineout
```

```python
name=BeautifulSoup(dineout.content)
```

```python
first_name=name.find("div",class_="restnt-info cursor")
first_name
```

```python
first_name.text
```

```python
title=[]
for i in name.find_all("div",class_="restnt-info cursor"):
    title.append(i.text)
title
```

```python
loc=name.find("div",class_="restnt-loc ellipsis")
loc
```

```python
loc.text
```

```python
location= []
for i in name.find_all("div",class_="restnt-loc ellipsis"):
    location.append(i.text)
location
 
```

```python
rating=name.find("div",class_="restnt-rating rating-4")
rating.text
```

```python
rating_all=[]
for i in name.find_all("div",class_="restnt-rating rating-4"):
    rating_all.append(i.text)
rating_all
```

```python
cuisine=name.find("span",class_="double-line-ellipsis")
cuisine.text
```

```python
cuisine_all=[]
for i in name.find_all("span",class_="double-line-ellipsis"):
    cuisine_all.append(i.text)
cuisine_all
```

```python
image=name.find("img",class_="no-img")
image
```

```python
images=[]
for i in name.find_all("img",class_="no-img"):
    images.append(i['data-src'])
images
```

```python
print(len(title),len(location),len(rating_all),len(cuisine_all),len(images))
```

```python
import pandas as pd 
```

```python
df= pd.DataFrame({"titles":title,"locations":location,"ratings":rating_all,"cuisine":cuisine_all,"images":images})
df
```
